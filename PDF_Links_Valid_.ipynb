```
from bs4 import BeautifulSoup
import pandas as pd
import requests
from habanero import Crossref
```

```
table = pd.read_excel('/PDF_List.xlsx')
```

Create list of PDF citations: (Copied from "Parse_Citations.ipynb")

```
titles = []
errors = []
DOI = []
i = 0
while(i < len(table)):
    print(i)
    soup = BeautifulSoup(table['Tables as HTML'][i], 'html5lib')
    if(soup.find("small", {"class":"references"}) == None or soup.find("small", {"class":"references"}) == ''):
        errors.append(i)
        titles.append("Error")
        DOI.append("Error")
        i = i + 1
        continue
    
    source = soup.find("small", {"class":"references"}).get_text()

    #remove numbers from citation (comes from post's citation list)
    
    if('[1]' in source and '[2]' not in source):
        reduced = source.replace('[1]','').strip()
        source = reduced
    elif('[2]' in source and '[1]' not in source):
        reduced = source.replace('[2]','').strip()
        source = reduced
    elif('[3]' in source and '[1]' not in source):
        reduced = source.replace('[3]','').strip()
        source = reduced   
    elif('[4]' in source and '[1]' not in source):
        reduced = source.replace('[4]','').strip()
        source = reduced
    elif('[5]' in source and '[1]' not in source):
        reduced = source.replace('[5]','').strip()
        source = reduced
    elif('[6]' in source and '[1]' not in source):
        reduced = source.replace('[6]','').strip()
        source = reduced
    elif('[7]' in source and '[1]' not in source):
        reduced = source.replace('[7]','').strip()
        source = reduced          

    if('1' in source):
      reduced = source.replace('1','').strip()
      source = reduced
    if('2' in source):
      reduced = source.replace('1','').strip()
      source = reduced
    if("\n" in source):
      reduced = source.replace("\n", " ")
      source = reduced
    if("  " in source):
      reduced = source.replace("  ", " ")
      source = reduced


    titles.append(source)
    cr = Crossref()
    x = cr.works(query = source, limit = 1)
    DOI.append([ z['DOI'] for z in x['message']['items'] ][0])
    i = i + 1
```

```
c = 0
while(c < len(table)):
  print(str(c) + ": " + titles[c])
  c = c + 1
table['titles'] = titles
print(table)
```

Check if PDF links reach actual PDFs:

```
count = 0;
while count < len(table['PDF']):
  print(count)  
  try:
    url = requests.get(table['PDF'][count])

  except requests.HTTPError as exception:
      print(exception)
      count = count + 1
      continue;
  
  soup = BeautifulSoup(url.content, 'html.parser')

  if soup.title is None:
    print("nonetype")
    count = count + 1
    continue;

  if ('search is temporarily unavailable' in soup.title.get_text()):
    print("error")
    continue;
    
  count = count + 1
  #firstDiv = soup.find(class_='div')
  #print(firstDiv)
```
